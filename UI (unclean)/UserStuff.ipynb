```python
import requests
from PIL import Image, UnidentifiedImageError
from io import BytesIO
import IPython.display as display
from bs4 import BeautifulSoup
import time

def is_valid_image(url, max_size=200):
    """Fetches and validates an image from a given URL, resizing it if necessary."""
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }
    try:
        response = requests.get(url, headers=headers, stream=True, timeout=5)
        response.raise_for_status()

        # Skip small/empty images
        if int(response.headers.get("Content-Length", 0)) < 2000:
            return None
        
        img = Image.open(BytesIO(response.content))

        # Check if the image is blank or transparent
        if img.getbbox() is None:
            return None

        # Resize image while maintaining aspect ratio
        if max(img.size) > max_size:
            img.thumbnail((max_size, max_size), Image.LANCZOS)

        return img

    except (requests.exceptions.RequestException, UnidentifiedImageError) as e:
        print(f" Request failed or image is corrupted: {e}")
        return None

def get_bing_image(isbn, book_title=None):
    """Scrapes Bing Images for an ISBN and optionally includes the book title in the search."""
    title_query = f"+{book_title.replace(' ', '+')}" if book_title else ""
    search_url = f"https://www.bing.com/images/search?q=ISBN+{isbn}{title_query}&form=HDRSC2"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }
    
    try:
        print(f" Searching Bing Images for ISBN: {isbn}{f' and Title: {book_title}' if book_title else ''}")
        
        response = requests.get(search_url, headers=headers, timeout=5)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find images that have the class "mimg"
        img_tags = soup.find_all("img", class_="mimg")

        for img in img_tags:
            img_src = img.get("src")

            # skip invalid images (placeholders, base64, or None)
            if img_src and img_src.startswith("https://tse"):
                return img_src

        print(" Bing search failed, no valid images found.")
        return None

    except requests.exceptions.RequestException as e:
        print(f" Bing search failed: {e}")
        return None

def fetch_and_display_image(row, max_size=200):
    """Fetches an image from a dataframe row and displays it along with the book title."""
    book_title = row.get("Book-Title")  # Default if title is missing
    print(f" Book Title: {book_title}")

    # Try Image-URL-L first
    if "Image-URL-L" in row and isinstance(row["Image-URL-L"], str):
        img = is_valid_image(row["Image-URL-L"], max_size)
        if img:
            display.display(img)
            return
    # then this url 
    if "Image-URL-M" in row and isinstance(row["Image-URL-M"], str):
        img = is_valid_image(row["Image-URL-M"], max_size)
        if img:
            display.display(img)
            return

    # If both fail, search Bing Images using ISBN and Title
    if "ISBN" in row and isinstance(row["ISBN"], str):
        bing_img_url = get_bing_image(row["ISBN"], book_title)
        if bing_img_url:
            img = is_valid_image(bing_img_url, max_size)
            if img:
                display.display(img)
                return

    print(" No valid image found.")
```
