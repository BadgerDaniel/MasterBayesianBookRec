```python
import requests
from PIL import Image
from io import BytesIO
import IPython.display as display

def show_image_from_url(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }
    
    try:
        response = requests.get(url, headers=headers, stream=True)
        response.raise_for_status()  
        
        img = Image.open(BytesIO(response.content))

        display.display(img)
        
    except requests.exceptions.RequestException as e:
        print(f"Error fetching image: {e}")
    except Exception as e:
        print(f"Error displaying image: {e}")

```

```python
import requests
from PIL import Image, UnidentifiedImageError
from io import BytesIO
import IPython.display as display
from bs4 import BeautifulSoup
import time


def is_valid_image(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }
    try:
        #print(f"Trying URL: {url}")  
        response = requests.get(url, headers=headers, stream=True, timeout=5)
        response.raise_for_status()

        # ensure file isn't too small and likely empty or placeholder
        if int(response.headers.get("Content-Length", 0)) < 2000:
            #print("Image is too small, likely empty. Skipping...")
            return None
        
        
        img = Image.open(BytesIO(response.content))

        # Check if the image is blank or transparent
        if img.getbbox() is None:  # Empty/transparent image check
            #print("Image appears to be blank. Skipping...")
            return None

        #print("Valid image found!")
        return img

    except (requests.exceptions.RequestException, UnidentifiedImageError) as e:
        print(f"Request failed or image is corrupted: {e}")
        return None

# Function to scrape Bing Images for ISBN, filtering only class "mimg"
def get_bing_image(isbn):
    search_url = f"https://www.bing.com/images/search?q=ISBN+{isbn}&form=HDRSC2"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }
    
    try:
        print(f" Searching Bing Images for ISBN: {isbn}")
        #print(f" URL: {search_url}")

        response = requests.get(search_url, headers=headers, timeout=5)
        response.raise_for_status()
        
        ## HTML PARSE
        soup = BeautifulSoup(response.text, 'html.parser')

        ### find the images that have the class "mimg"
        img_tags = soup.find_all("img", class_="mimg")
        #print(f"🔢 Found {len(img_tags)} image tags with class 'mimg'")

        for img in img_tags:
            img_src = img.get("src")
            
            # debug code, to show images
            #print(f" Found Image URL: {img_src}")

            # Skip invalid images (placeholders, base64, or None)
            if img_src and img_src.startswith("https://tse"):
                #print(f"✔ Using Bing Image: {img_src}")
                return img_src

        print(" Bing search failed")
        return None
    except requests.exceptions.RequestException as e:
        print(f"⚠ Bing search failed: {e}")
        return None

## final function to fetch and display the image from df or Bing
def fetch_and_display_image(row):
    # Try Image-URL-L first
    if "Image-URL-L" in row and isinstance(row["Image-URL-L"], str):
        img = is_valid_image(row["Image-URL-L"])
        if img:
            display.display(img)
            return

    # Image-URL-M if Image-URL-L fails
    if "Image-URL-M" in row and isinstance(row["Image-URL-M"], str):
        img = is_valid_image(row["Image-URL-M"])
        if img:
            display.display(img)
            return

    # if both fail, search Bing Images using ISBN
    if "ISBN" in row and isinstance(row["ISBN"], str):
        bing_img_url = get_bing_image(row["ISBN"])
        if bing_img_url:
            img = is_valid_image(bing_img_url)
            if img:
                display.display(img)
                return

    print("❌ No valid image found.")


fetch_and_display_image(random_users_books_t3.iloc[0])  
```
